
(d) Drop any columns that are missing more than 70% of their data.

```python
for column in train:
  if train[column].isna().sum()/train[column].count() > 0.7:
    print(f'{column} has more than 70% of data is na')
    train = train.drop(column, axis=1)
```

```python
train
```

(e) Replace missing values with the mean value of the existing data for that feature. If it is a categorical variable, replace it by the mode.

```python
train = train.fillna(train.mean())
```

```python
train = train.fillna(train.mode())
```

```python
train.isna().sum().sum()
```


(f) Extract the numerical data only. We will only use those for regression for the time being. We will not use the categorical valued features.

```python
numer_cols = train.loc[:, train.dtypes != 'object'].columns
```

```python
numer_cols
```

```python
train = train[numer_cols]
```

```python
train
```

(g) Choose the 10 best features based on their linear relationship with the target variable in SalePrice. This would be based on how strongly they correlate with SalePrice.

```python
features = train.corr()['SalePrice'].sort_values(ascending=False)[1:11].index
features
```


(h) Create a correlation matrix with the top 10 features.

```python
import seaborn as sns
import matplotlib.pyplot as plt
correlation_matrix = train[features].corr(method='pearson').round(2) # 2 decimal accuracy
sns.heatmap(data=correlation_matrix, annot=True)
plt.title('Pearson pair-wise Correlation Matrix')
plt.show()
```

(i) Remove features of the subset of the best features from the above step that are co-linear.
Here, we will use the metric for co-linearity as highly correlated (≥ 0.8 correlation).

```python
features = features.drop([features[2], features[4], features[7]])
```

```python
train[features]
```

(k) Run multiple linear regression on this training set with the pertinent features, report R2 and RMSE.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

x = train[features]
y = np.log(train['SalePrice'])

lr_model = LinearRegression()
lr_model.fit(x, y)
```

```python
lr_model.intercept_
```

```python
lr_model.coef_
```

```python
y_predict = lr_model.predict(x)
mean_squared_error(y, y_predict)
```

```python
r2_score(y, y_predict)
```

```python
sum(abs(y_predict-y))
```

(l) Evaluate (not train) your multiple linear regression model on the testing set with the pertinent features selected earlier.

```python
x_test = test[features]
y_test_predict = lr_model.predict(x_test)
```

(n) Use polynomial regression on the selected features and re-train your model. Do any of the metrics improve over the linear one?

```python
# yes, both r2 and rmse are improved
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=4)
x_poly = poly_features.fit_transform(x)

lrp_model = LinearRegression()
lrp_model.fit(x_poly, y)
```

```python
x_poly.shape
```

```python
y_poly_predict = lrp_model.predict(x_poly)
mean_squared_error(y, y_poly_predict)
```

```python
r2_score(y, y_poly_predict)
```
